# Labeler And Label Saver Functions

The following function allows me to manually label as much data as I want, so that I can do some semi-structured learning.

**Labels = `{1: Neg, 2: Neut, 3: Pos}`**

```python
def save_labels(df):
    '''
    Basic function to save labels outside of notebook for easy access later.
    '''
    with open('labels/labels.txt', 'w') as fp:
        json.dump(list(df['sentiment']), fp)
```

```python
def label_sentiment(df, count=100):
    '''
    Iterates through a DataFrame and randomly selects an unlabeled tweet to
    show the user and get manual sentiment label for. Sentiment label is saved
    to DataFrame.
    
    Sub out df['tweet'] to make this function applicable to any other labeling
    task.
    
    Input 'x' as label to break out of function early.
    ---
    df
        What DataFrame you are labeling in.
    count
        Defaults to 100. Number of tweets to label.
    '''
#     Show full tweets when printing. No truncation.
    pd.set_option('display.max_colwidth', -1)
    
    for i in range(count):
        
#     Get random row that hasn't been labeled yet.
        row_to_label = df[df['sentiment'] == 'unlabeled'].sample()
    
#     Show row to user and store their sentiment label of given tweet.
        print(row_to_label['tweet'])
        label = input("Sentiment? ")
        if label == 'x':
            return

#     Store index of shown tweet and relabel sentiment at stored index.
        row_index = int(row_to_label.index.values)
        df.iloc[row_index]['sentiment'] = label
        
#     Clunky, but saving every iteration incase I accidentally close it
        save_labels(df)
```

# Labeling Data

***A note on criteria for sentiment:*** Word choice is more important than overall tweet purpose in certain cases. Many tweets about or at Amazon are just product promotions. In theory, these could all be classed as positive, since they promote the brand. Many of these tweets will be labeled as such, but some (such as ad tweets about horror movies or murder mystery novels) which are laden with negative words, will be classed as negative.

Tweets made by actual people and not just a brand or product account follow more conventional definitions of sentiment, referring simply to how the author likely was feeling. Moreover, I'd rather have a neural net that understands word choice than one that understands what is or isn't promotional content. No sense in reaching out to accounts that already promote.
